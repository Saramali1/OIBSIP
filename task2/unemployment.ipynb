**UNEMPLOYMENT ANALYSIS WITH PYTHON**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Unemployment in India.csv')
```

```python
print(df.columns)
```

```python
df.rename(columns={
    ' Estimated Unemployment Rate (%)': 'Unemployment Rate',
    'Region': 'Region',
    ' Date': 'Date',
    ' Frequency': 'Frequency',
    ' Estimated Employed': 'Employed',
    ' Estimated Labour Participation Rate (%)': 'Labour Participation Rate',
    'Area': 'Area'
}, inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
low_threshold = 3.0 
high_threshold = 7.0
```

```python
df['Unemployment Category'] = pd.cut(df['Unemployment Rate'], 
bins=[0, low_threshold, high_threshold, df['Unemployment Rate'].max()],
labels=['Low Unemployment', 'Moderate Unemployment', 'High Unemployment'])
```

```python
category_counts = df['Unemployment Category'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
```

```python
labels = category_counts.index
sizes = category_counts.values
colors = ['#99ff99', '#c2c2f0', '#ff9999']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('Unemployment Categories')

plt.axis('equal') 

plt.tight_layout()
plt.show()
```

```python
unique_regions = df['Region'].unique()
```

```python
for region in unique_regions:
    plt.figure(figsize=(8, 4))
```

```python
region_data = df[df['Region'] == region]
```

```python
plt.hist(region_data['Unemployment Rate'], bins=10, alpha=0.7)
    
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.title(f'Unemployment Rate Histogram for {region}')
    
plt.tight_layout()
plt.show()
```
